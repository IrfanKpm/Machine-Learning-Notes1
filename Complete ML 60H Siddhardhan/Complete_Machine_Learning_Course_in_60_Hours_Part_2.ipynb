```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create a basic dataset
data = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50])
df = pd.DataFrame(data, columns=['Values'])

df
```

```python

# Calculate mean, median, mode
mean = df['Values'].mean()
median = df['Values'].median()

# Calculate range
data_range = df['Values'].max() - df['Values'].min()

# Calculate variance and standard deviation
variance = df['Values'].var()
std_dev = df['Values'].std()

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Range: {data_range}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

```python
# Generate a dataset with 1200 points from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=15, size=1200)  # Mean=50, StdDev=15
df = pd.DataFrame(data, columns=['Values'])

# Adding noise to a range of points
# Create a range of points and add noise
range_points = np.linspace(df['Values'].min(), df['Values'].max(), 300)
noise = np.random.normal(loc=0, scale=5, size=300)  # Noise with standard deviation of 5
noisy_range_points = range_points + noise

# Append noisy range points to the dataframe
df = pd.concat([df, pd.DataFrame(noisy_range_points, columns=['Values'])], ignore_index=True)

# Recalculate mean and standard deviation
mean = df['Values'].mean()
std_dev = df['Values'].std()

# Plot standard deviation
plt.figure(figsize=(8, 5))
sns.histplot(df['Values'], kde=True, color='blue')
plt.axvline(mean, color='red', linestyle='--', label=f'Mean ({mean:.2f})')
plt.axvline(mean - std_dev, color='green', linestyle='--', label=f'SD - ({mean - std_dev:.2f})')
plt.axvline(mean + std_dev, color='green', linestyle='--', label=f'SD + ({mean + std_dev:.2f})')
plt.title('Standard Deviation Plot')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Normalize data
df['Normalized'] = (df['Values'] - df['Values'].min()) / (df['Values'].max() - df['Values'].min())

# Plot before and after normalization
plt.figure(figsize=(11, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Values'], kde=True, color='blue')
plt.title('Before Normalization')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df['Normalized'], kde=True, color='green')
plt.title('After Normalization')
plt.xlabel('Normalized Values')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
