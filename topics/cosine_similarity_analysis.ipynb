### **Overview of cosine_similarity_analysis**

#### Key Features:

1. **TF-IDF Vectorization:**
   - Converts a set of textual phrases into numerical vectors using Term Frequency-Inverse Document Frequency (TF-IDF). This representation captures the importance of each term within the context of the phrases, allowing for meaningful comparisons.

2. **Cosine Similarity and Distance Computation:**
   - Computes the cosine similarity and distance between all pairs of phrases. Cosine similarity measures how closely related two vectors are by calculating the cosine of the angle between them, while cosine distance represents the dissimilarity.

3. **Visualization:**
   - Provides heatmaps to visually represent the cosine similarity and distance matrices. These heatmaps help in easily identifying the relationships and distances between different topics, making it simpler to understand their connections.

4. **Data Presentation:**
   - Displays the TF-IDF values for each phrase in a structured format, allowing for detailed inspection of how each term contributes to the vector representation of the phrases.

#### Applications:

- **Text Analysis:**
  - Useful for analyzing and comparing the similarity between various topics or documents in fields such as content analysis, information retrieval, and natural language processing.

- **Educational Insights:**
  - Helps in understanding how closely related different educational or technical topics are, which can aid in curriculum design or content organization.

- **Recommendation Systems:**
  - Can be utilized in developing recommendation systems by assessing the similarity between different items or topics to suggest relevant content.

The `cosine_similarity_analysis` script for exploring the semantic relationships between textual data, providing both quantitative metrics and visual insights to enhance understanding and analysis.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances
import pandas as pd
```

```python
# Define the phrases (video topics)
phrases = [
    "Introduction to C++ Programming", "C++ Vector Data Structures", "Python Object Oriented Programming",
    "Using Scapy for Python", "Basics of Machine Learning", "Deep Learning with TensorFlow",
    "Fundamentals of Data Science", "Artificial Intelligence Overview", "Developing with Web Technologies",
    "Django Web Framework Tutorial", "Building APIs with Flask", "TensorFlow Basics for Beginners",
    "Keras for Deep Learning", "Handling Data with Pandas", "NumPy Array Operations",
    "Writing SQL Database Queries", "JavaScript Programming Fundamentals", "HTML and CSS Basics",
    "React JavaScript Framework", "Node.js Server Programming"
]
```

```python
# Convert phrases to TF-IDF vectors
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(phrases)
```

```python
# Convert TF-IDF matrix to DataFrame for better readability
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out(), index=phrases)

tfidf_df.head(15)
```

```python
# Print the TF-IDF values
print("TF-IDF Values:")
print(tfidf_df)
```

```python
# Calculate cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix)
cosine_dist = cosine_distances(tfidf_matrix)
```

```python
# Plotting the cosine similarity
plt.figure(figsize=(12, 10))
sns.heatmap(cosine_sim, annot=True, cmap="coolwarm", xticklabels=phrases, yticklabels=phrases)
plt.title("Cosine Similarity")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```

```python
# Plotting the cosine distance
plt.figure(figsize=(12, 10))
sns.heatmap(cosine_dist, annot=True, cmap="coolwarm", xticklabels=phrases, yticklabels=phrases)
plt.title("Cosine Distance")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances
from sklearn.decomposition import PCA
import plotly.express as px
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Define the phrases (video topics)
phrases = [
    "Introduction to C++ Programming", "C++ Vector Data Structures", "Python Object Oriented Programming",
    "Using Scapy for Python", "Basics of Machine Learning", "Deep Learning with TensorFlow",
    "Fundamentals of Data Science", "Artificial Intelligence Overview", "Developing with Web Technologies",
    "Django Web Framework Tutorial", "Building APIs with Flask", "TensorFlow Basics for Beginners",
    "Keras for Deep Learning", "Handling Data with Pandas", "NumPy Array Operations",
    "Writing SQL Database Queries", "JavaScript Programming Fundamentals", "HTML and CSS Basics",
    "React JavaScript Framework", "Node.js Server Programming"
]

# Convert phrases to TF-IDF vectors
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(phrases)

# Convert TF-IDF matrix to DataFrame for better readability
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out(), index=phrases)

# Print the TF-IDF values
print("TF-IDF Values:")
print(tfidf_df)

# Calculate cosine similarity and distance
cosine_sim = cosine_similarity(tfidf_matrix)
cosine_dist = cosine_distances(tfidf_matrix)

# Plotting the cosine similarity with Plotly for interactivity
fig_sim = px.imshow(cosine_sim, text_auto=True, labels=dict(x="Phrases", y="Phrases", color="Cosine Similarity"),
                    title="Cosine Similarity Heatmap")
fig_sim.update_xaxes(tickangle=45)
fig_sim.update_yaxes(tickangle=0)
fig_sim.show()

# Plotting the cosine distance with Plotly for interactivity
fig_dist = px.imshow(cosine_dist, text_auto=True, labels=dict(x="Phrases", y="Phrases", color="Cosine Distance"),
                     title="Cosine Distance Heatmap")
fig_dist.update_xaxes(tickangle=45)
fig_dist.update_yaxes(tickangle=0)
fig_dist.show()

# Dimensionality Reduction using PCA
scaler = StandardScaler()
tfidf_scaled = scaler.fit_transform(tfidf_matrix.toarray())
pca = PCA(n_components=2)
pca_result = pca.fit_transform(tfidf_scaled)

# Plotting PCA results
plt.figure(figsize=(10, 8))
plt.scatter(pca_result[:, 0], pca_result[:, 1])
for i, phrase in enumerate(phrases):
    plt.text(pca_result[i, 0], pca_result[i, 1], phrase)
plt.title("PCA of TF-IDF Vectors")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid()
plt.show()

# Dimensionality Reduction using t-SNE
tsne = TSNE(n_components=2, random_state=0)
tsne_result = tsne.fit_transform(tfidf_scaled)

# Plotting t-SNE results
plt.figure(figsize=(10, 8))
plt.scatter(tsne_result[:, 0], tsne_result[:, 1])
for i, phrase in enumerate(phrases):
    plt.text(tsne_result[i, 0], tsne_result[i, 1], phrase)
plt.title("t-SNE of TF-IDF Vectors")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.grid()
plt.show()

# Clustering using K-Means
kmeans = KMeans(n_clusters=4, random_state=0)
clusters = kmeans.fit_predict(tfidf_matrix)

# Plotting clusters with t-SNE
plt.figure(figsize=(10, 8))
scatter = plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=clusters, cmap='viridis')
plt.colorbar(scatter, label='Cluster')
for i, phrase in enumerate(phrases):
    plt.text(tsne_result[i, 0], tsne_result[i, 1], phrase)
plt.title("t-SNE Clustering of TF-IDF Vectors")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.grid()
plt.show()
```
