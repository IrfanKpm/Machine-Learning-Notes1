```python
from PIL import Image
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
def plot_img(img):
  plt.figure(figsize=(4,4))
  plt.imshow(img)
  plt.show()
```

## **PIL (Pillow)**
Pillow is a powerful and easy-to-use Python library for image processing. It is a modern fork of the Python Imaging Library (PIL), and it provides a comprehensive suite of tools for opening, manipulating, and saving various image formats. Pillow supports numerous image file formats including JPEG, PNG, GIF, BMP, and TIFF, among others.

```python
print(np.array(r)) # red chanel of img2
```

```python
# Create a figure and a 2x2 grid of subplots
fig, ax = plt.subplots(2, 2, figsize=(14, 10))

# Plot each channel in its respective subplot
ax[0, 0].imshow(r, cmap='Reds')
ax[0, 0].set_title('Red Channel')
ax[0, 0].axis('off')  # Hide the axis

ax[0, 1].imshow(g, cmap='Greens')
ax[0, 1].set_title('Green Channel')
ax[0, 1].axis('off')  # Hide the axis

ax[1, 0].imshow(b, cmap='Blues')
ax[1, 0].set_title('Blue Channel')
ax[1, 0].axis('off')  # Hide the axis

ax[1, 1].imshow(a, cmap='gray')
ax[1, 1].set_title('Alpha Channel')
ax[1, 1].axis('off')  # Hide the axis

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
