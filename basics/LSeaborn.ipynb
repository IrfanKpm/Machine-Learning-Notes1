##**Introduction to Seaborn**

Welcome to the Seaborn tutorial notebook, designed to introduce you to Seaborn, a powerful Python visualization library based on Matplotlib.

##**What is Seaborn?**

Seaborn is a Python library for statistical data visualization built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

##**Uses of Seaborn**

**Statistical Visualization:** Seaborn excels in creating informative and visually appealing statistical graphics.

**Default Aesthetics:** It offers aesthetically pleasing default themes and color palettes for various plot types.

**Integration with Pandas:** Seaborn seamlessly integrates with Pandas DataFrames, making it easy to visualize data from data manipulation workflows.

```python
# Use sns.scatterplot() to create a scatter plot.
# This function plots data points on a 2D plane where x-axis represents 'tip' and y-axis represents 'total_bill'.

# Syntax:
# sns.scatterplot(x="tip", y="total_bill", data=tips)

# Parameters:
# - x: Specifies the data for the x-axis. Here, it represents the 'tip' column from the 'tips' DataFrame.
# - y: Specifies the data for the y-axis. Here, it represents the 'total_bill' column from the 'tips' DataFrame.
# - data: The DataFrame containing the data to be plotted. In this case, it's the 'tips' DataFrame.

# Example:
sns.scatterplot(x="tip", y="total_bill", data=tips)
```

```python
# Use sns.scatterplot() to create a scatter plot with additional features.
# Each point is colored based on the categorical variable specified by the 'hue' parameter, and sized based on the numerical variable specified by the 'size' parameter.



# Parameters:
# - x: Specifies the data for the x-axis. Here, it represents the 'tip' column from the 'tips' DataFrame.
# - y: Specifies the data for the y-axis. Here, it represents the 'total_bill' column from the 'tips' DataFrame.
# - data: The DataFrame containing the data to be plotted. In this case, it's the 'tips' DataFrame.
# - hue: Optional. Specifies a categorical variable in the DataFrame to use for coloring the data points. Here, it represents the 'day' column.
# - size: Optional. Specifies a numerical variable in the DataFrame to use for sizing the data points. Here, it represents the 'size' column.

# Example:
sns.scatterplot(x="tip",y="total_bill",data=tips,hue="day",size="size")
```

```python
# Use sns.histplot() to create a histogram plot.
# This function visualizes the distribution of numerical data by binning it into intervals (bins) and displaying the count or density of observations within each bin.

# Syntax:
# sns.histplot(data, x=None, y=None, hue=None, bins='auto', kde=False, stat='count', common_bins=True, common_norm=True, log_scale=False, fill=True)

# Parameters:
# - data: DataFrame, array, or list. Input data for plotting.
# - x, y: Variables in the data. Specify either x or y. If using DataFrame, these should be column names.
# - hue: Optional. Grouping variable that will produce points with different colors. Can be categorical or numeric.
# - bins: Specification of hist bins. 'auto', integer, or sequence of scalars. Default is 'auto' which uses Sturges' formula.
# - kde: Whether to plot a kernel density estimate (smoothed histogram). Default is False.
# - stat: Statistic to compute within each bin. 'count' for count of data points, 'frequency' for normalized count (density). Default is 'count'.
# - common_bins: Whether to use the same bins for all plots. Default is True.
# - common_norm: Whether to ensure all histograms are normalized to the same height. Default is True.
# - log_scale: Whether to plot the histogram with a logarithmic scale on the y-axis. Default is False.
# - fill: Whether to fill the bars of the histogram. Default is True.

# Example:
sns.histplot(data=tips, x='tip', bins=20, kde=True, stat='density', hue='day', log_scale=False, fill=True)
```

```python
# Use sns.barplot() to create a bar plot.
# This function displays the mean (or another estimator) of a numerical variable grouped by one or two categorical variables.

# Syntax:
# sns.barplot(x=None, y=None, hue=None, data=None, estimator=<function mean>, ci=95, n_boot=1000, units=None, seed=None, orient=None, color=None, palette=None, saturation=0.75, errcolor='.26', errwidth=None, capsize=None, dodge=True, ax=None, **kwargs)

# Parameters:
# - x, y: Variables in the data. Specify x and y for horizontal or vertical bar plots respectively.
# - hue: Optional. Grouping variable that will produce bars with different colors. Can be categorical.
# - data: DataFrame or array. Input data for plotting.
# - estimator: Function that aggregates data within each categorical bin. Default is mean.
# - ci: Confidence interval size for error bars. Default is 95.
# - palette: Optional. Color palette for the plot. Choose from seaborn palettes or provide a list of colors.
# - dodge: Whether to separate bars for different hues along the categorical axis. Default is True.

# Example:
sns.barplot(x="sex", y="tip", data=tips, hue='sex', palette="YlGnBu")

# Explanation:
# - x: Categorical variable 'sex' is plotted on the x-axis.
# - y: Numerical variable 'tip' is plotted on the y-axis, showing the mean tip amount.
# - hue: Variable 'sex' is used to group data by gender, with bars colored accordingly.
# - palette: Color palette "YlGnBu" is used to color the bars.
```

```python
# Use sns.stripplot() to create a strip plot.
# This function draws a scatter plot where one variable is categorical. It shows all individual data points along with a distribution of values.

# Syntax:
# sns.stripplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, jitter=True, dodge=False, palette=None, size=5, edgecolor='gray', linewidth=0.5, alpha=1.0, marker='o', s=None)

# Parameters:
# - x, y: Variables in the data. Specify x and y for horizontal or vertical strip plots respectively.
# - hue: Optional. Grouping variable that will produce points with different colors. Can be categorical.
# - data: DataFrame or array. Input data for plotting.
# - order, hue_order: Lists of strings. Specify the order of categorical levels.
# - jitter: Amount of jitter (random noise) applied to data points along the categorical axis to avoid overlap. Default is True.
# - dodge: Whether to separate points for different hues along the categorical axis. Default is False.
# - palette: Optional. Color palette for the plot. Choose from seaborn palettes or provide a list of colors.
# - size: Size of the markers (data points). Default is 5.
# - edgecolor: Color of marker edges. Default is 'gray'.
# - linewidth: Width of marker edges. Default is 0.5.
# - alpha: Transparency of markers. Default is 1.0 (fully opaque).
# - marker: Marker style. Default is 'o' (circle).
# - s: Optional. Size of the markers in points^2. Overrides 'size' parameter.

# Example:
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu')

# Explanation:
# - x: Categorical variable 'day' is plotted on the x-axis.
# - y: Numerical variable 'tip' is plotted on the y-axis, showing individual tip amounts.
# - hue: Variable 'sex' is used to group data by gender, with points colored accordingly.
# - jitter: Adds random jitter to data points along the categorical axis to avoid overlap.
# - dodge: Does not separate points for different hues (sex) along the categorical axis in this example.
# - palette: Color palette "YlGnBu" is used to color the points.
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue='sex',palette="YlGnBu",dodge=True) # saperate sex
```

```python
# Use sns.jointplot() to create a joint plot with regression.
# This function combines scatter plots and histograms in a single figure, along with a regression line (kind='reg').

# Syntax:
# sns.jointplot(x=None, y=None, data=None, kind='scatter', color=None, height=6, ratio=5, space=0.2, dropna=True, xlim=None, ylim=None, marginal_ticks=False, hue=None, palette=None, hue_order=None, bins='auto', linewidth=1, edgecolor='w', annot=None, **kwargs)

# Parameters:
# - x, y: Variables in the data. Specify x and y for the scatter plot.
# - data: DataFrame or array. Input data for plotting.
# - kind: Type of plot to draw. Use 'reg' for regression plot along with scatter and histograms.
# - height: Height of the plot in inches.
# - ratio: Ratio of joint axes height to marginal axes height.
# - space: Space between the joint and marginal axes.
# - xlim, ylim: Limits for the x and y axes.
# - hue: Optional. Grouping variable that will produce points with different colors. Can be categorical.
# - palette: Optional. Color palette for the plot.
# - bins: Number of bins to use in histograms. Default is 'auto'.
# - linewidth, edgecolor: Styling parameters for scatter plot markers.
# - annot: Optional annotation function or boolean for adding annotations to the plot.

# Example:
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')

```

```python
# Syntax:
# sns.jointplot(x=None, y=None, data=None, kind='scatter', color=None, height=6, ratio=5, space=0.2, dropna=True, xlim=None, ylim=None, marginal_ticks=False, hue=None, palette=None, hue_order=None, bins='auto', linewidth=1, edgecolor='w', annot=None, **kwargs)

# Parameters:
# - x, y: Variables in the data. Specify x and y for the scatter plot.
# - data: DataFrame or array. Input data for plotting.
# - kind: Type of plot to draw. Use 'kde' for kernel density estimation plot along with scatter and histograms.
# - height: Height of the plot in inches.
# - ratio: Ratio of joint axes height to marginal axes height.
# - space: Space between the joint and marginal axes.
# - xlim, ylim: Limits for the x and y axes.
# - hue: Optional. Grouping variable that will produce points with different colors. Can be categorical.
# - palette: Optional. Color palette for the plot.
# - bins: Number of bins to use in histograms. Default is 'auto'.
# - linewidth, edgecolor: Styling parameters for scatter plot markers.
# - annot: Optional annotation function or boolean for adding annotations to the plot.

# Example:
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde')


```

```python
# Use sns.jointplot() to create a joint plot with kernel density estimation (KDE).
# Set 'kind' to 'kde' and 'fill' to True to shade the KDE contours.

# Syntax:
# sns.jointplot(x=None, y=None, data=None, kind='scatter', color=None, height=6, ratio=5, space=0.2, dropna=True, xlim=None, ylim=None, marginal_ticks=False, hue=None, palette=None, hue_order=None, bins='auto', linewidth=1, edgecolor='w', annot=None, **kwargs)

# Parameters:
# - x, y: Variables in the data. Specify x and y for the scatter plot.
# - data: DataFrame or array. Input data for plotting.
# - kind: Type of plot to draw. Use 'kde' for kernel density estimation plot along with scatter and histograms.
# - height: Height of the plot in inches.
# - ratio: Ratio of joint axes height to marginal axes height.
# - space: Space between the joint and marginal axes.
# - xlim, ylim: Limits for the x and y axes.
# - hue: Optional. Grouping variable that will produce points with different colors. Can be categorical.
# - palette: Optional. Color palette for the plot.
# - bins: Number of bins to use in histograms. Default is 'auto'.
# - linewidth, edgecolor: Styling parameters for scatter plot markers.
# - annot: Optional annotation function or boolean for adding annotations to the plot.
# - shade: Optional. If True, shade the KDE contours.

# Example:
sns.jointplot(x='tip',y='total_bill',data=tips,kind='kde',fill=True)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='kde',fill=True,cmap="YlGnBu")  # cmap :: colormap
```
