```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4,5,6,7,8]
y = [0,2,4,6,8,10,12,14,16]
y1 = [2,5,1,14,9,12,16,4,1]

plt.plot(x,y,c="green",lw=3,label='2x')
plt.plot(x,y1,c="red",lw=2,label='random',linestyle="--")
plt.title("Our First Graph!")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.xticks([0,1.5,3,4.5,6,7.5,])
plt.yticks([0,3,6,9,12,15])

plt.legend()

plt.show()
```

```python
x = np.random.random(200) * 100
y = np.random.random(200) * 100
plt.figure(figsize=(4,3))
plt.scatter(x,y,c="red",marker="*",s=70,alpha=0.6) # s > size
plt.show()
```

```python
x = np.linspace(0,2*np.pi,1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10,5))
plt.plot(x, y1,c="red" ,label='sin(x)')
plt.plot(x, y2,c="green" ,label='cos(x)')

plt.title("Graph of Sin(x) and Cos(x)")
plt.xlabel("x")
plt.ylabel("func(x)")
plt.legend()

plt.grid(True)
plt.show()
```

```python
x = ["c++","c#","Python","Java","Go"]
y = [30,50,140,10,45]
plt.figure(figsize=(5,4))
plt.bar(x,y,color="purple",edgecolor="red",lw=2)
plt.show()
```

```python
x = np.random.normal(20,1,1000)
plt.hist(x,bins=20)
plt.show()
```

```python
x = ["c++","c#","Python","Java","Go"]
y = [30,50,140,10,45]
explodes = [0,0,0,0.07,0]
plt.figure(figsize=(5,4))
plt.pie(y,labels=x,explode=explodes,autopct="%.2f%%")
plt.show()
```

```python
heights = np.random.normal(172,8,300)
plt.boxplot(heights)
plt.show()
```

```python
x1 , y1 = np.random.random(300) , np.random.random(300)
x2 , y2 = np.arange(100) , np.random.random(100)

plt.figure(1,figsize=(5,4))
plt.scatter(x1,y1,c="g",alpha=0.6)

plt.figure(2,figsize=(5,4))
plt.plot(x2,y2)

plt.show()
```

```python
x = np.arange(100)
fig , axs = plt.subplots(2,2, figsize=(9, 7))

axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title("sin wave")

axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title("cos wave")

axs[1,0].plot(x,np.random.random(100))
axs[1,0].set_title("random function")

axs[1,1].scatter(x,np.random.random(100))
axs[1,1].set_title("random scatter")

plt.show()
```
